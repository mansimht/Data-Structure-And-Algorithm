**What is tuple?**

- A tuple is sequence of values like a list.
- A value store in tuple can be anytype and they are indexed by integers
- A tuple is immutable sequence of python object. Once we declare that we cannot change them
- Tuples are also comparable and hashable. (Hashable means object is set to be hashable if the value remains same during lifetime.)


```python
#how to create tuple
new_tuple=('a','b','c','d','e')
print(new_tuple)

new_tuple2= ('a',)
print(new_tuple2)

new_tuple3= ('a')
print(new_tuple3)

new_tuple4= tuple('abcde')
print(new_tuple4)

new_tuple5= tuple(['a','b','c','d','e'])
print(new_tuple5)

#Time complexity: O(1)
#Space Complexity: O(n)
```

```python
#how can access element of tuple?
new_tuple=('a','b','c','d','e')
print(new_tuple[1])
print(new_tuple[-1])
print(new_tuple[::-1])
print(new_tuple[:])
print(new_tuple[3:])
```

```python
#Traversing tuple
new_tuple=('a','b','c','d','e')
for i in new_tuple:
  print(i)

for i in range(len(new_tuple)):
  print(new_tuple[i])

#Time complexity: O(n)
#Space complexity: O(1)
```

```python
#Search for element in tuple
new_tuple=('a','b','c','d','e')
print('a' in new_tuple)
print('f' in new_tuple)
#Time complexity: linear_search: O(n)
#Space complexity: O(1)

print(new_tuple.index('c'))
#Time complexity: O(n)
#Space complexity: O(1)

def new_search(tuples,element):
  for i in range(len(tuples)):
    if tuples[i]==element:
      return f"The {element} is found at index {i}"
  return 'The element is not found'

print(new_search(new_tuple,'c'))
print(new_search(new_tuple,'h'))
#Time complexity: O(n)
#Space complexity: O(1)
```

```python
#Tuple operations and methods
tuple1= (1,4,3,2,5,2,4,2)
tuple2= (1,2,6,9,8,7)

print(tuple1+tuple2)  #Concatenation of tuple
print(tuple1 * 5)   #star operator with tuple and integer
print(tuple1.count(2))   #count method
print(tuple1.index(5))   #index method
print(len(tuple1))   #len method
print(max(tuple1))   #max method
print(min(tuple1))   #min method
print(sum(tuple1))   #sum method
```

```python
#Tuple is immutable and List ia mutable
lst= [0,1,2,3,4,5,6]
print(lst)
lst[4]= 10
print(lst)
del lst[1]
print(lst)

lst1= [(0,1),(2,3),(4,5)]
print(lst1)

tuple1= (0,1,2,3,4,5,6)
print(tuple1)
tuple1=  7,8,9,10
print(tuple1)
del tuple1
print(tuple1)

#Tuples are used for hetrogeneous(different)datatype and list are use for homogenous(same) datatype
#Iterating tuple is faster than list
#Tuple that contain immutable elements can be used as key in dictionary
#If you have data doesn't change, implementing it as tuple it will guarantee that it remains protected
#Tuples can be stored in lists and lists can be stored in tuple
#both tuples and list can be nested
```

**Time and Space Complexity**

1. Creating a tuple:
   - Time Complexity: O(1)
   - Space Complexity: O(n)

2. Traversing tuple:
   - Time Complexity: O(n)
   - Space Complexity: O(1)

3. Accessing given element:
   - Time Complexity: O(1)
   - Space Complexity: O(1)

4. Searching given element:
   - Time Complexity: O(n)
   - Space Complexity: O(1)
