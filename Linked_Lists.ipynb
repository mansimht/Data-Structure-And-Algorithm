**What is Linked Lists**

- Linked lists is a form of sequential collection and it does not have to be in order.
- A linked list is made up of independent nodes that may contain any type of data and each has refeerence to the next node in the link.

Ex. Train- where cars(nodes) and connected with joints(links)
         - which has head(train engine) and tail(last car)
         - each car is indepent
         - Each element consists 2 parts(nodes: data and links)
         - each car is linked to the next car


- Nodes: data and refrences
- Head for locating inked lists in memory. Then we have nodes over here, which are the elements of linke lists.

- Node consists 2 parts:
  1. Value of node and
  2. Refrence to next node which creates link between curret and next node




          

**Linked List**

- Linked list doesn't have indexes
- They are not contigious with each other(element not located to each other)
- Here first node head and last node tail, and first node point to second node,second node to third node and then to last node
- Nodes are spread out t different places in memory and then this nodes are coneected to each other via pointers.

**List**

- lists have indexes
- list elements are contigious with each other
- Access element using indexes


**Types of Linked List**

1. Singly Linked Lists
2. Circular Linked Lists
3. Doubly Linked Lists
4. Circular Doubly Linked Lists

**Singly Linked Lists**
- In single linked list each node in the lists stores the value and refrences to the next node in the linked list.
- It doesnot have any refrence to previous node
2 Parts: first part is value and second is refrence(physical address of second node)

**Circular Linked lists**
- It is same as singly linked lists. The only difference is last node of this list stores refrence to the first node.
- In singly linked list last node is always null but here last node have refrence to first node.
- **Ex.** Chess with 4 player

**Doubly linked lists**
- This type of linked list also resembles a single linked list.
- The diffrence we have two refrences from each node refrence to previous node and refrence to next node.
- We can move forward and backward
- **Ex.** Music app we change next and previous song

**Circular Doubly linked list**
- Last node we have refrences to next node.
- Physcial location of first node is stored in next refrence of last node and physical location o last node is stored in previous refrence of first node.
- **Ex.** In mac cmd+shift+Tab= to switch app infinelty(App1,App2,App3,App4)





**Linked List in memory**

- There is nodes in momory where first node(1) is pointed by head, second node(2) is pointed by first node(1), third node(4) is pointed by second node(2), fourth node(5) is pointed by third node(4) and fourth node(5) is pointed by tail node.
- If create new node in memory and linked to avialable node.
- Linked lists allowd us dynamically sizing at runtime and because of random location we cannot access any given element directly
- So we have to traverse from first element to find the required element which is one of the weak points of linked lists
- So by knowing location of one, we cannot add a value to the index of this node and find the next node in the memory. Because next node can be anywhere in the memory.
