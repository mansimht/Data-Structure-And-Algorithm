```python
#Accessing and traversing list
shoppingList= ['Milk','Cheese','Butter']

#Accessing list
print(shoppingList[0])
print(shoppingList[1])
print(shoppingList[2])
print('Milk' in shoppingList)
print('Bread' in shoppingList)
print(shoppingList[-1])
print(shoppingList[:-1])
print(shoppingList[::-1])

#Traversing List
for shopping in shoppingList:
  print(shopping)

for i in range(len(shoppingList)):
  shoppingList[i] = shoppingList[i] + '+'
  print(shoppingList[i])
```

```python
#Update and Insert List
lst= []

#Update method
lst.append(1)
lst.append(2)
lst.append(3)          #Time complxity= O(1),#Space complexity= O(1)
lst.append(4)
print(lst)

lst.insert(4,7)        #Time complxity= O(n),#Space complexity= O(n)
print(lst)

lst.extend([6,7,8])     #Time complxity= O(n),#Space complexity= O(n)
print(lst)

#==================================================
#insert method
lst[2]= 12
print(lst)

#Time complxity= O(1)
#Space complexity= O(1)  # no extra time and space required
```

```python
#Slice and Delete element from list


mylst= ['a','b','c','d','e','f']

#Slicing in list=[:]
print(mylst[0:2]) #first 2 element
print(mylst[-1]) #last element
print(mylst[:])  #all element
print(mylst[::-1])  #reverse element
print(mylst[::2])  #alternate element
print(mylst[1::2])  #alternate element starting from index 1
print(mylst[1:4:2])  #alternate element starting from index 1 to 4

#Time complexity= O(1)
#Space complexity= O(1)

#Deleting element of list
mylst.pop(2)         #Time complxity= O(n) and if delete last element then O(1),#Space complexity= O(1)
print(mylst)

mylst.remove('e')     #Time complxity= O(n),#Space complexity= O(1)
print(mylst)

del mylst[0]        #Time complxity= O(n),#Space complexity= O(1)
print(mylst)
del mylst[2:4]
print(mylst)
```

```python
#Searching element in list
lst1= [10,20,30,40,50,60,70,80,90]

# in operator
target= 500
if target in lst1:
  print(f"Target {target} found in list")
else:
  print(f"Target {target} not found in list")

#Time complexity= O(n) in worst case= as item perform  search to compare,
#Space complexity= O(1)

# linear search

def linear_search(lst,target):
  for i in range(len(lst)):
    if lst[i] == target:
      return i
  return -1

print(linear_search(lst1,60))
print(linear_search(lst1,700))

#Time complexity= O(n)
#Space complexity= O(1)
```

**List v/s arrays**

**Similarities**

1. Both data structures are mutable
2. Both can be indexed and interated through
3. They can be both sliced

**Difference**

1. Arrays are for arithmetic oprations
2. List can add different data type

```python
import numpy as np

arr= np.array([1,2,3,4,5,6])
lst= [1,2,3,4,5]

print(arr/2)
print(lst/2)      # list cannot perform airthmetic operation like list
```

**Time and Space complexity**

1. Creating an empty list:
   - Time Complexity: O(1)     - only intialize empty list no data
   - Space Complexity: O(1)    - no additional memory required

2. Creating List with elements:
   - Time Complexity: O(n)     - only intialize empty list add data with n elements
   - Space Complexity: O(n)    - additional memory required as per n element

3. Inserting value in a list:
   - Time Complexity: O(n)     - all element shift right or left as per n element
   - Space Complexity: O(1)    - no additional memory required

4. Traversing List:
   - Time Complexity: O(n)     - all element traaverse as per n element
   - Space Complexity: O(1)    - no additional memory required

5. Accessing a given cell in list:
   - Time Complexity: O(1)     - access element via indexes
   - Space Complexity: O(1)    - no additional memory required

6. Searching a given value in List:
   - Time Complexity: O(n)     - linear search take O(n) time complexity
   - Space Complexity: O(1)    - no additional memory required

7. Deleting a value from list:
   - Time Complexity: O(n)     - delete first element and all element will shift left
   - Space Complexity: O(1)    - no additional memory required

```python
#example1
data= [[[1,2],[3,4]],[5,6],[[7,8]]]
def fun(m):
  v = m[0][0]
  for row in m:
    for element in row:
      if v < element:
        v = element
  return v
print(fun(data[0]))
```
