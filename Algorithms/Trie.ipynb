**what is Trie?**

- A trie is tree based data structure tht organizes information in a heirarchy.
- Trie is used for storing words in a way which enables fast lookups.
- SInce tire stores characters at each node, its very efficient prefix matching in english language

**Properties**
- It is typically ised to store or search strings in a space and time efficient way
- ANy node in try can store non repitative multiple characters
- Every node stores link of next character of the string
- Every node keeps track of end of the string

```
      AB
     / \
    I  AIM
   /   /  \
  RT   R   L
```
AIR,AIT,BAR,BIL,BM= is trie strucutre
- AIR: First create one node as root node and store A in it, then create child node of it and stor I, Then subchild of A which store R and last blank node will created store end of string
- BAR: In root node B is also store at there, now right subchild node store A and their subchild left node store R and then new left sub child node store end of string
- BIL: In root node B is also store at there, now right subchild node store I with A and their subchild left node store  L and then new left sub child node store end of string
- BM: In root node B is also store at there, now right subchild node store M and then new left sub child node store end of string

**Searching in Trie**
- Case1: String doesn't exist in Trie
- Case2: String exist in Trie

````python
#Searching in Trie
class Trienode:
  def __init__(self):
    self.child= {}
    self.endofstring= False

class Trie:
  def __init__(self):
    self.root= Trienode()

  def insert(self,word):
    current= self.root
    for i in word:
      ch= i
      node= current.child.get(ch)
      if node == None:
        node= Trienode()
        current.child.update({ch:node})
      current= node
    current.endofstring = True
    print("Word of string successfully Inserted")

  def search(self,word):
    current= self.root
    for i in word:
      ch= i
      node= current.child.get(ch)
      if node == None:
        return False
      current= node
    if current.endofstring == True:
      return True
    else:
      return False


newtrie= Trie()
newtrie.insert("App")
newtrie.insert("Appl")
print(newtrie.search('App'))
print(newtrie.search('Ap'))

#Time complexity:O(n)
#Space Complexity: O(1)
````
