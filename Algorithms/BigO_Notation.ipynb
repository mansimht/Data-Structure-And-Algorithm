```python
## Drop Constants= means easily eliminate constant values from asymptotic analysis
def print_items1(n):
  for i in range(n):
    print(i)
  for j in range(n):
    print(j)

print_items1(10)

# here n+n= O(2n) time complexity

# It is very possible that O(N) code is faster than the O(1) code for specific inputs.

# Different Computers with different architechtures have different constant factors
# ex. fast computer,fast memory access but lower constant and slow computer,slow memory access and higher constant

# Different algorithm with same basic idea and computational complexity might have different constants
# ex. a*(b-c) vs a*b-a*c
# a*(b-c) have lower constant then a*b-a*c

# in asymtotic analysis, we ignore constant factors because it removes a lot og noise from hadware and small change in algorithm to make easy algorithm
```

```python
##Drop Non dominant features

def print_items1(n):
  for i in range(n):
    for j in range(n):       # O(N^2)
      print(i,j)
  for k in range(n):         # O(N)
    print(k)

print_items1(3)

# Total time complexity O(n^2 + n)
# n^2 is dominat term and which is higher then n non-dominant term so we will drop non dominant feature.
# So total Time Complexity is O(n^2)
```

**Space Complexity:**

- Memory space required by algorithm

- Howmuch memory in the worst case is needed in algorithm.
Also need to take case how memory size grows with input grows

```python
def sum(n):
  if n <= 0:
    return 0
  return n + sum(n-1)

print(sum(3))

# meomory space complexity is O(n)
#first call sum(3)== in stack store sum(3)
#then recursively call sum(2) so stack store sum(2)
#then recursively call sum(1) so stack store sum(1) continue when n<0.
```

```python
def pair_sum_sequence(n):
  total=0
  for i in range(n):
    total= total + pair_sum(i,i+1)
  return total

def pair_sum(a,b):
  return a+b

print(pair_sum_sequence(3))

# O(1) space complexity
```

```python
# different term for input- add vs multiply

def print_items1(a,b):
  for i in range(a):     # O(a)
    print(i)
  for j in range(b):     # O(b)
    print(j)

print_items1(2,2)


# time complexity is O(a+b)
# space complexity is O(1)
```

**How to measure the code using Big O**

- Rule1: If any assignents and if statements are executed once regardless of size of problem===> O(1)
- Rule2: A simple for loop from 0 to n(no internal loops)===> O(n)
- Rule3: A nested loop of same type take quadratic time complexity===> O(n^2)
- Rule4: A loop, in which controlling paramter is devided by two at each step(binary search or devide and conquer)===> O(logN)
- Rule5: When dealing with multiple statment, just add them up===> O(N+M) or O(N*M)
