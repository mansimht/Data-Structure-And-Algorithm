**What is AVL Tree**

- AVL tree is self balancing BST where the difference between heights of left and right subtree cannot be more than one for all nodes

- If at any time heights if left and right subtree differ by ore than one, then rebalancing is done to restore AVL property, this process is called rotation


**Why AVL Tree?**

- Ex. 10,20,30,40,50,60,70
- Here as BST, if we inserted this nodes so all nodes are inserted at right side only.
- If we need to search 60, then we need to visit all nodes. SO time complexity is O(n), because of height is more.
- BST height is lower than this tree and time complexity for searching is O(logN).
- AVL trees make sure that we never end up with this kind of disbalanced over here.
- AVL tree enforces some rules when we are inserting or deleting nodes from binary search tree to make sure that tree will be never disbalanced
- BST can get squeezed and hence its performance, thus going down.
Instead of O(logN) for insertion, searching, and deleting it can go up to O(N) time complexity
- To prevent such cases AVL tree help us to balance the binary search tree and boost its performance of insertion, searching, and deleting
- Every time tree get disbalanced, AVL tree prform some rotation to get tree balanced again

**Insertions and Balancing:**
- Insert 30 → Root node.
- Insert 25 → Left of 30.
- Insert 35 → Right of 30.
- Insert 20 → Left of 25.
- Insert 15 → Left of 20.
- Insert 5 → Left of 15.
  Imbalance at 25 (left-heavy) → Right rotation at 25.
- Insert 10 → Right of 5.
- Insert 50 → Right of 35.
- Insert 60 → Right of 50.
- Insert 70 → Right of 60.
  Imbalance at 35 (right-heavy) → Left rotation at 35.

  Imbalance at 60 (right-heavy) → Left rotation at 60.

````python
#insert a nodein AVL
from collections import deque

class AVLnode:
  def __init__(self,data):
    self.data= data
    self.left= None
    self.right= None
    self.height=1

def levelorder(root):
  if not root:
    return
  q= deque()
  q.append(root)
  while q:
    size= len(q)
    for i in range(size):
      node= q.popleft()
      print(node.data)
      if node.left:
        q.append(node.left)
      if node.right:
        q.append(node.right)

#Insertion helper function
#get height of root node
def getheight(root):
  if not root:
    return 0
  return root.height

#rotation right
def rotateright(disbalancednode):
  newroot= disbalancednode.left
  disbalancednode.left= disbalancednode.left.right
  newroot.right= disbalancednode
  disbalancednode.height = 1 + max(getheight(disbalancednode.left),getheight(disbalancednode.right))
  newroot.height= 1 + max(getheight(newroot.left),getheight(newroot.right))
  return newroot

#rotate left
def rotateleft(disbalancednode):
  newroot= disbalancednode.right
  disbalancednode.right= disbalancednode.right.left
  newroot= disbalancednode
  disbalancednode.height = 1 + max(getheight(disbalancednode.left),getheight(disbalancednode.right))
  newroot.height= 1 + max(getheight(newroot.left),getheight(newroot.right))
  return newroot

#checking balance left and riht subtree
def getbalance(root):
  if not root:
    return 0
  return getheight(root.left)-getheight(root.right)

def insert(root,newnode):
  if not root:
    return AVLnode(newnode)
  elif newnode < root.data:
    root.left= insert(root.left,newnode)
  elif newnode > root.data:
    root.right = insert(root.right,newnode)

  root.height= 1 + max(getheight(root.left),getheight(root.right))
  #get balanced of rootnode
  balance= getbalance(root)
  #eft left condition= right rotation
  if balance > 1 and newnode < root.left.data:
    return rotateright(root)
  #left eight condition= left right rotation
  if balance > 1 and newnode > root.left.data:
    root.left= rotateleft(root.left)
    return rotateright(root)
  #right right condition= left rotation
  if balance > 1 and newnode > root.right.data:
    return rotateleft(root)
  #right left condition= right left condition
  if balance > 1 and newnode < root.right.data:
    root.right= rotateright(root.right)
    return rotateleft(root)
  return root


newAVL= AVLnode(5)
newAVL= insert(newAVL,10)
newAVL= insert(newAVL,50)
newAVL= insert(newAVL,20)
levelorder(newAVL)

#Time Complexity: O(logN)
#Space Complexity: O(logN)
````

**Time and Space Complexity of AVL Tree**

1. Create AVL:
   - Time compelxity: O(1)
   - Space complexity: O(1)

2. Insert a node AVL:
   - Time compelxity: O(logn)
   - Space complexity: O(logn)

3. Traverse AVL:
   - Time compelxity: O(n)
   - Space complexity: O(n)

4. Search for node AVL:
   - Time compelxity: O(logn)
   - Space complexity: O(logn)

5. Delete node from AVL:
   - Time compelxity: O(logn)
   - Space complexity: O(logn)

6. Delete Entire AVL:
   - Time compelxity: O(1)
   - Space complexity: O(1)
