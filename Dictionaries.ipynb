**Dictonary in memory**

- Dictonary implement using hashtable
- hashtable is way of doing key value lookups.
- You store value in array and use a hash function to find the index of array cell that corresponding to key-value pair.
- This improve performance and minimize number of collison(different keys has same values and index)  


```python
#Insert/Update element in memory

mydict= {'name':'Edy','age':26}
print(mydict)
mydict['age']=32
print(mydict)
mydict['address']='London'
print(mydict)

#Time complexity: O(1)
#Space complexity: O(1)
```

```python
#Traverse through dictionary
my_dict= {'name': 'Edy', 'age': 32, 'address': 'London'}

for k,v in my_dict.items():
  print(k,v)

for key in my_dict:
  print(key,my_dict[key])

#Time complexity= O(n)
#Space Complexity= O(1)
```

```python
#search element in dictionary= linear search
my_dict= {'name': 'Edy', 'age': 32, 'address': 'London'}

def searchele(dict,value):
  for key in dict:
    if dict[key] == value:
      return key,value
  return 'not found'

print(searchele(my_dict,'London'))
print(searchele(my_dict,'Ahmedabad'))
#Time complexity= O(n)
#Space Complexity= O(1)
```

```python
#Delete remove element of array
my_dict= {'name': 'Edy', 'age': 32, 'address': 'London'}
print(my_dict)
del my_dict['age']
print(my_dict)
#Time complexity: O(1)
#Space complexity: O(1)

my_dict= {'name': 'Edy', 'age': 32, 'address': 'London'}
remove_element= my_dict.pop('name')
print(remove_element)
print(my_dict)
#Time complexity: O(1)
#Space complexity: O(1)


my_dict= {'name': 'Edy', 'age': 32, 'address': 'London'}
remove_element= my_dict.popitem()
print(remove_element)
#Time complexity: O(1)
#Space complexity: O(1)

my_dict= {'name': 'Edy', 'age': 32, 'address': 'London'}
my_dict.clear()
print(my_dict)
#Time complexity: O(n)
#Space complexity: O(1)
```

**Dictionary methods**

- clear()= delete all method
- copy()= copy of dictionary and not modify orignal one
- fromkeys()= create new dictonary from given sequence of elements with value provided by user
- get()= return values of specify key if key in dictonary
- items()= return all key value pair
- keys()= return list of all keys
- popitem()= remove last key-value pair from dictonary
- setdefault()= return value of key if key is in dictonary, if not it insert key with value
- pop()= remove element key-value
- values()=return all values
- update()= update key-value in dictonary or add key-value



```python
my_dict= {'name': 'Edy', 'age': 32, 'address': 'London'}
dict1= my_dict.copy()
print(dict1)

dict2= my_dict.fromkeys(dict1)
print(dict2)

dict3= my_dict.get('name')
print(dict3)

dict4= my_dict.items()
print(dict4)

dict5= my_dict.keys()
print(dict5)

dict6= my_dict.popitem()
print(my_dict)

dict7= my_dict.setdefault('name1','Mansi')
print(my_dict)

dict8= my_dict.pop('name1','Mansi')
print(my_dict)

dict9= my_dict.values()
print(dict9)

dict10= my_dict.update({'name':'Mansi'})
print(my_dict)
```

```python
#Dictonary operations

my_dict= {4: 'Four',3: 'Three',5: 'Five'}
print(3 in my_dict)
print('Three' in my_dict)
print('Three' in my_dict.values())
print(len(my_dict))
print(sorted(my_dict))
print(all(my_dict))
print(any(my_dict))
```

**Time and Space complexity in Dictionary**

1. Create a dictionary:
   - Time Complexity: O(len(dict))
   - Space complexity: O(n)

2. Inserting a value:
   - Time Complexity: O(1)/O(n)
   - Space complexity: O(1)

3. Traversing given dictionary:
   - Time Complexity: O(n)
   - Space complexity: O(1)
  
4. Accessing a given cell:
   - Time Complexity: O(1)
   - Space complexity: O(1)

5. Searching given value:
   - Time Complexity: O(n)-linear search and in operator-O(1)
   - Space complexity: O(1)

6. Deleting given value:
   - Time Complexity: O(1)
   - Space complexity: O(1)
